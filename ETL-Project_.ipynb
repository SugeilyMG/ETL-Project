### Importing Dependencies

```python
import pandas as pd
import numpy
```

### Import CSV into Jupyter


```python
governors = "Resources/US Elections 2020/governors_county_candidate.csv"
president = "Resources/US Elections 2020/president_county_candidate.csv"
senate = "Resources/US Elections 2020/senate_county_candidate.csv"
socialmedia = "Resources/Congress on Social Media/Social_media_statistics.csv"
top100associated = "Resources/Congress on Social Media/Top_100_terms_associated.csv"
```

### Explore CSV's and Transform the Data
#### ⦁	Review the dataframes to know the fields and define the columns we will need for the analysis of each table.
#### ⦁	Change the name of the columns to match all the tables
#### ⦁	Define the fields  we will we used to join the tables and in this case we change it to match in all the tables
#### ⦁	Filter the information necessary for the analysis.

#### I.	Transform Governors Dataframe

```python
##Editing dataframe; renaming columns, replacing values...

governors_df = pd.read_csv(governors)
govparty_df = governors_df.rename(columns={"party":"Party"})
govparty_df["Party"] = govparty_df["Party"].replace({"DEM": "D", "REP":"R"})
govparty_df.head()
```

```python
##Filtering data for Democrats and Republicans
##Final df 

govfinal_df = govparty_df.loc[(govparty_df['Party']== 'D') | (govparty_df['Party']== 'R')]
govfinal_df.head(10)
```

#### II. Transform President Dataframe

```python
##Editing dataframe; renaming columns, replacing values...

president_df = pd.read_csv(president)
preparty_df = president_df.rename(columns={"party":"Party","total_votes":"votes"})
preparty_df["Party"] = preparty_df["Party"].replace({"DEM": "D", "REP":"R"})
preparty_df.head()
```

```python
##Filtering data for Democrats and Republicans
##Final df 

prefinal_df = preparty_df.loc[(preparty_df['Party']== 'D') | (preparty_df['Party']== 'R')]
prefinal_df.head(10)
```

#### III. Transform Senate Dataframe

```python
## senate: Editing dataframe; renaming columns, replacing values...

senate_df = pd.read_csv(senate)
senparty_df = senate_df.rename(columns={"party":"Party","total_votes":"votes"})
senparty_df["Party"] = senparty_df["Party"].replace({"DEM": "D", "REP":"R"})
senparty_df.head()
```

```python
##Filtering data for Democrats and Republicans
##Final df 

senfinal_df = senparty_df.loc[(senparty_df['Party']== 'D') | (senparty_df['Party']== 'R')]
senfinal_df.head(10)
```

#### IIII. Transform Social Media Dataframe

```python
##Printing df 

socialmedia_df = pd.read_csv(socialmedia)
socialmedia_df.head()
```

```python
##Unique values

socialmedia_df["Party"].unique()
```

```python
##Unique values

socialmedia_df["First Post"].unique()
```

```python
##Editing dataframe; filtering data for Democrats and Republicans

socialmedia_df2 =socialmedia_df
socialmedia_df2['Party']=socialmedia_df2["Party"].replace({"(R)": "R", "(D)":"D"})
socialmedia_df2.head()
```

```python
##Editing dataframe; dropping columns (inplace method)
##Printing df

socialmedia_df2.drop(["Bioguide ID","Number of Active Accounts","Start of Term",\
                                  "End of Term", "First Post", "Last Post"], axis=1, inplace=True)
socialmedia_df2.head(10)
```

```python
socialmedia_df3 = socialmedia_df2.rename(columns={"Max Total Followers":"Max_Followers",
                                                  "Total Posts":"Total_Posts",
                                                 "Average Post Favorites/Reactions":"Average_FavReac",
                                                 "Average Post Retweets/Shares":"Average_RetweetsShares"})
```

```python
socialmedia_df3.head(10)
```

#### IIIII. Transform top100associated Dataframe

```python
##Printing df

top100associated_df = pd.read_csv(top100associated)
top100associated_df.head()
```

```python
##Unique values

top100associated_df["Party"].unique()
```

```python
##Unique values

top100associated_df["Year"].unique()
```

```python
##Editing dataframe; dropping columns (inplace method)
##Printing df

top100associated_df2 = top100associated_df.loc[(top100associated_df['Year']== 2019) | (top100associated_df['Party']== 2020)]
top100associated_df2.head(10)
```
